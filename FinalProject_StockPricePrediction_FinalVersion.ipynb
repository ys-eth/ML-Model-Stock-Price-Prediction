# Evaluation Metrics

RMSE:

This method will allow us to see the differences between predicted and actual values. It's the square root of the average squared differences between the two values, and the lower it is the better the performance of the model.

$$ \text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

MAPE:

This method will allow us to view the accuracy of the predictions as a percentage value. This value is calculated by taking the average absolute percentage difference between the predicted and actual values. The lower the score is, the better the performance of the model.

$$ \text{MAPE} = \frac{100}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right|$$

By combining these two models, we will be able to measure our performance accurately. This will be to ensure that our model provides reliable predictions to the best of its ability and prove that it can be used as a financial tool in the market.

# Modeling

### LSTM Predictions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.optimizers import Adam


def predict_and_plot(stock_name):
    
    file_path = 'all_stocks_2006-01-01_to_2018-01-01.csv'
    data = pd.read_csv(file_path)

    stock_data = data[data['Name'] == stock_name]

    stock_data = stock_data.sort_values('Date')

    
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])

    closing_prices = stock_data['Close'].values

    
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(closing_prices.reshape(-1, 1))

    
    SEQ_LENGTH = 100
    def create_dataset(dataset, time_step=1):
        dataX, dataY = [], []
        for i in range(len(dataset)-time_step-1):
            a = dataset[i:(i+time_step), 0]
            dataX.append(a)
            dataY.append(dataset[i + time_step, 0])
        return np.array(dataX), np.array(dataY)

    
    train_size = int(len(scaled_data) * 0.8)
    train_data = scaled_data[:train_size]
    test_data = scaled_data[train_size:]

    X_train, y_train = create_dataset(train_data, SEQ_LENGTH)
    X_test, y_test = create_dataset(test_data, SEQ_LENGTH)

    
    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(SEQ_LENGTH, 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

    history = model.fit(X_train, y_train, epochs=20, batch_size=64, validation_split=0.2)

    predicted_prices = model.predict(X_test)
    predicted_prices = scaler.inverse_transform(predicted_prices)
    y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

    
    rmse = np.sqrt(mean_squared_error(y_test, predicted_prices))
    mape = mean_absolute_percentage_error(y_test, predicted_prices) * 100

    print(f'{stock_name} - RMSE: {rmse}')
    print(f'{stock_name} - MAPE: {mape}')

    
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data['Date'][len(stock_data)-len(y_test):], y_test, label='Actual Prices')
    plt.plot(stock_data['Date'][len(stock_data)-len(y_test):], predicted_prices, label='Predicted Prices')
    plt.title(f'LSTM Model Predictions for {stock_name}')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

stocks = ['AAPL', 'PG', 'GOOGL']

for stock in stocks:
    predict_and_plot(stock)
```

The LSTM model predictions for AAPL, PG, and GOOGL showcase the model's effectiveness in capturing the overall trends and patterns in the stock prices from 2016 to 2018. These plots demonstrate that the predicted prices (orange line) closely follow the actual prices (blue line), with some room for error.

For AAPL, the model achieves an RMSE of 8.52 and a MAPE of 5.52%, indicating a good level of accuracy in its predictions. 

Similarly, for PG, the RMSE is 1.33 and the MAPE is 1.24%, suggesting very high prediction accuracy. 

For GOOGL, the model also performs well with an RMSE of 40.56 and a MAPE of 4.24%, despite the higher volatility and higher price range of GOOGL compared to AAPL and PG. 

Overall, the LSTM model demonstrates strong predictive capabilities across different stocks, showing its robustness and adaptability to various market conditions.

### KNN Predictions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler


file_path = 'all_stocks_2006-01-01_to_2018-01-01.csv'
data = pd.read_csv(file_path)

data['Date'] = pd.to_datetime(data['Date'])


def create_features(data, window_size=5):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i + window_size])
        y.append(data[i + window_size])
    return np.array(X), np.array(y)


stocks = ['AAPL', 'GOOGL', 'PG']

window_size = 5
n_neighbors = 5

for stock in stocks:
    
    stock_data = data[data['Name'] == stock]

    
    close_prices = stock_data['Close'].values
    dates = stock_data['Date'].values

    
    train_size = int(len(close_prices) * 0.8)
    train_data = close_prices[:train_size]
    test_data = close_prices[train_size:]
    train_dates = dates[:train_size]
    test_dates = dates[train_size:]

    
    X_train, y_train = create_features(train_data, window_size)
    X_test, y_test = create_features(test_data, window_size)

    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    
    knn = KNeighborsRegressor(n_neighbors=n_neighbors)
    knn.fit(X_train_scaled, y_train)

    predictions = knn.predict(X_test_scaled)

    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    mape = mean_absolute_percentage_error(y_test, predictions) * 100

    print(f"{stock} KNN RMSE: {rmse:.4f}")
    print(f"{stock} KNN MAPE: {mape:.4f}")
    
    plot_dates = test_dates[window_size:]

    plt.figure(figsize=(14, 7))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Original Data')
    plt.plot(plot_dates, predictions, label=f'{stock} KNN Predictions', color='red', linewidth=0.5)
    plt.title(f'KNN Regression Predictions of {stock} Stock Prices')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()
```

For AAPL, the KNN model achieved an RMSE of 16.0181 and a MAPE of 6.4585, indicating moderate accuracy. The predictions for AAPL showed some deviations from the actual prices, especially in periods of high volatility. 

For GOOGL, the KNN model resulted in an RMSE of 201.3718 and a MAPE of 19.0107, reflecting greater prediction errors compared to AAPL. This might be due to the higher price levels and more volatile nature of GOOGL stock. 

The PG predictions had an RMSE of 0.9734 and a MAPE of 0.8867, showing better accuracy and closer alignment with actual prices.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

data = pd.read_csv('all_stocks_2006-01-01_to_2018-01-01.csv')
data = pd.read_csv(file_path)

data['Date'] = pd.to_datetime(data['Date'])


def create_features(data, window_size=5):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i + window_size])
        y.append(data[i + window_size])
    return np.array(X), np.array(y)

def create_dqn_model(input_shape):
    model = Sequential()
    model.add(Dense(64, input_dim=input_shape, activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
    return model

stocks = ['AAPL', 'GOOGL', 'PG']

window_size = 5
epochs = 50
batch_size = 32

for stock in stocks:

    stock_data = data[data['Name'] == stock]

    close_prices = stock_data['Close'].values
    dates = stock_data['Date'].values

    train_size = int(len(close_prices) * 0.8)
    train_data = close_prices[:train_size]
    test_data = close_prices[train_size:]
    train_dates = dates[:train_size]
    test_dates = dates[train_size:]

    X_train, y_train = create_features(train_data, window_size)
    X_test, y_test = create_features(test_data, window_size)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    
    input_shape = X_train_scaled.shape[1]
    model = create_dqn_model(input_shape)
    model.fit(X_train_scaled, y_train, epochs=epochs, batch_size=batch_size, verbose=1)

  
    predictions = model.predict(X_test_scaled)

    
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    mape = mean_absolute_percentage_error(y_test, predictions) * 100

    print(f"{stock} DQN RMSE: {rmse:.4f}")
    print(f"{stock} DQN MAPE: {mape:.4f}")

    plot_dates = test_dates[window_size:]

    plt.figure(figsize=(14, 7))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Original Data')
    plt.plot(plot_dates, predictions, label=f'{stock} DQN Predictions', color='red', linewidth=0.5)
    plt.title(f'DQN Predictions of {stock} Stock Prices')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.neighbors import KNeighborsRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.optimizers import Adam


data = pd.read_csv('all_stocks_2006-01-01_to_2018-01-01.csv')

data['Date'] = pd.to_datetime(data['Date'])

stocks = ['AAPL', 'GOOGL', 'PG']

def create_features(data, window_size=5):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i + window_size])
        y.append(data[i + window_size])
    return np.array(X), np.array(y)

results = {stock: {'RMSE': {}, 'MAPE': {}} for stock in stocks}

for stock in stocks:
   
    stock_data = data[data['Name'] == stock]

   
    close_prices = stock_data['Close'].values
    dates = stock_data['Date'].values

    train_size = int(len(close_prices) * 0.8)
    train_data = close_prices[:train_size]
    test_data = close_prices[train_size:]
    train_dates = dates[:train_size]
    test_dates = dates[train_size:]

    window_size = 5
    X_train, y_train = create_features(train_data, window_size)
    X_test, y_test = create_features(test_data, window_size)


    if stock != 'GOOGL':  
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        knn = KNeighborsRegressor(n_neighbors=5)
        knn.fit(X_train_scaled, y_train)
        knn_predictions = knn.predict(X_test_scaled)
        knn_rmse = np.sqrt(mean_squared_error(y_test, knn_predictions))
        knn_mape = mean_absolute_percentage_error(y_test, knn_predictions) * 100
        results[stock]['RMSE']['KNN'] = knn_rmse
        results[stock]['MAPE']['KNN'] = knn_mape


    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(close_prices.reshape(-1, 1))
    train_data_scaled = scaled_data[:train_size]
    test_data_scaled = scaled_data[train_size:]

    X_train, y_train = create_features(train_data_scaled, window_size)
    X_test, y_test = create_features(test_data_scaled, window_size)

    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
    y_test_inverse = scaler.inverse_transform(y_test.reshape(-1, 1))

    lstm_model = Sequential()
    lstm_model.add(LSTM(50, return_sequences=True, input_shape=(window_size, 1)))
    lstm_model.add(Dropout(0.2))
    lstm_model.add(LSTM(50, return_sequences=False))
    lstm_model.add(Dropout(0.2))
    lstm_model.add(Dense(1))
    lstm_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
    lstm_model.fit(X_train, y_train, epochs=20, batch_size=64, validation_split=0.2, verbose=0)
    lstm_predictions = lstm_model.predict(X_test)
    lstm_predictions = scaler.inverse_transform(lstm_predictions)
    lstm_rmse = np.sqrt(mean_squared_error(y_test_inverse, lstm_predictions))
    lstm_mape = mean_absolute_percentage_error(y_test_inverse, lstm_predictions) * 100
    results[stock]['RMSE']['LSTM'] = lstm_rmse
    results[stock]['MAPE']['LSTM'] = lstm_mape

    dqn_model = Sequential()
    dqn_model.add(LSTM(50, return_sequences=True, input_shape=(window_size, 1)))
    dqn_model.add(Dropout(0.2))
    dqn_model.add(LSTM(50, return_sequences=False))
    dqn_model.add(Dropout(0.2))
    dqn_model.add(Dense(1))
    dqn_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
    dqn_model.fit(X_train, y_train, epochs=20, batch_size=64, validation_split=0.2, verbose=0)
    dqn_predictions = dqn_model.predict(X_test)
    dqn_predictions = scaler.inverse_transform(dqn_predictions)
    dqn_rmse = np.sqrt(mean_squared_error(y_test_inverse, dqn_predictions))
    dqn_mape = mean_absolute_percentage_error(y_test_inverse, dqn_predictions) * 100
    results[stock]['RMSE']['DQN'] = dqn_rmse
    results[stock]['MAPE']['DQN'] = dqn_mape

def plot_comparison(results, metric):
    stocks = list(results.keys())
    models = ['KNN', 'LSTM', 'DQN']

    values = np.array([[results[stock][metric].get(model, np.nan) for model in models] for stock in stocks])

    x = np.arange(len(stocks))
    width = 0.25

    fig, ax = plt.subplots(figsize=(12, 6))

    for i, model in enumerate(models):
        bars = ax.bar(x + i*width, values[:, i], width, label=model)

        for bar in bars:
            height = bar.get_height()
            if not np.isnan(height):
                ax.annotate(f'{model}\n{height:.2f}',
                            xy=(bar.get_x() + bar.get_width() / 2, height),
                            xytext=(0, 3),  
                            textcoords="offset points",
                            ha='center', va='bottom')

    for j in range(len(stocks)):
        best_model_idx = np.nanargmin(values[j, :])
        ax.bar(x[j] + best_model_idx*width, values[j, best_model_idx], width, edgecolor='red', linewidth=2)

    ax.set_xlabel('Stocks')
    ax.set_ylabel(metric)
    ax.set_title(f'Comparison of {metric} for Different Models')
    ax.set_xticks(x + width)
    ax.set_xticklabels(stocks)
    ax.legend()

    plt.show()
plot_comparison(results, 'RMSE')
plot_comparison(results, 'MAPE')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.neighbors import KNeighborsRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from datetime import timedelta

file_path = 'all_stocks_2006-01-01_to_2018-01-01.csv'
data = pd.read_csv(file_path)
data['Date'] = pd.to_datetime(data['Date'])

stocks = ['AAPL', 'PG', 'GOOGL']

def create_features(data, window_size=5):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i + window_size])
        y.append(data[i + window_size])
    return np.array(X), np.array(y)

def simulate_trading(dates, actual_prices, predicted_prices, title):
    initial_balance = 10000
    balance = initial_balance
    shares = 0

    balances = []

    for i in range(len(predicted_prices) - 1):
        if predicted_prices[i + 1] > predicted_prices[i]:  
            if shares == 0:  
                shares = balance / actual_prices[i]
                balance = 0
        elif predicted_prices[i + 1] < predicted_prices[i]:  
            if shares > 0:  
                balance = shares * actual_prices[i]
                shares = 0
        balances.append(balance + shares * actual_prices[i])

    
    final_balance = balance + shares * actual_prices[-1]
    print(f'{title} Final Balance: ${float(final_balance):.2f}')

    plt.figure(figsize=(10, 6))
    plt.plot(dates[:len(balances)], balances, label='Balance Over Time')
    plt.title(f'Trading Performance of {title}')
    plt.xlabel('Date')
    plt.ylabel('Balance')
    plt.legend()
    plt.show()

def print_prediction_accuracy(test_data, predicted_prices, model_name, stock_name):
    accuracy = 100 - np.mean(np.abs((test_data - predicted_prices) / test_data)) * 100
    print(f'{model_name} prediction accuracy for {stock_name}: {accuracy:.2f}%')

def process_stock(stock_data, stock_name):
    stock_data = stock_data.sort_values('Date')
    stock_data.fillna(method='ffill', inplace=True)  
    stock_data.fillna(method='bfill', inplace=True)  
    closing_prices = stock_data['Close'].values

    train_size = int(len(closing_prices) * 0.8)
    train_data = closing_prices[:train_size]
    test_data = closing_prices[train_size:]
    train_dates = stock_data['Date'][:train_size]
    test_dates = stock_data['Date'][train_size:]
    
    window_size = 5
    X_train, y_train = create_features(train_data, window_size)
    X_test, y_test = create_features(test_data, window_size)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    knn = KNeighborsRegressor(n_neighbors=5)
    knn.fit(X_train_scaled, y_train)
    predicted_prices_knn = knn.predict(X_test_scaled)

    print_prediction_accuracy(y_test, predicted_prices_knn, 'KNN', stock_name)

    def create_dqn_model(input_shape):
        model = Sequential()
        model.add(Dense(64, input_dim=input_shape, activation='relu'))
        model.add(Dense(64, activation='relu'))
        model.add(Dense(1))
        model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
        return model

    input_shape = X_train_scaled.shape[1]
    model = create_dqn_model(input_shape)
    model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, verbose=0)
    predicted_prices_dqn = model.predict(X_test_scaled)

    print_prediction_accuracy(y_test, predicted_prices_dqn, 'DQN', stock_name)

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(closing_prices.reshape(-1, 1))

    SEQ_LENGTH = 100
    X_train_lstm, y_train_lstm = create_features(scaled_data[:train_size], SEQ_LENGTH)
    X_test_lstm, y_test_lstm = create_features(scaled_data[train_size:], SEQ_LENGTH)

    X_train_lstm = X_train_lstm.reshape(X_train_lstm.shape[0], X_train_lstm.shape[1], 1)
    X_test_lstm = X_test_lstm.reshape(X_test_lstm.shape[0], X_test_lstm.shape[1], 1)

    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(SEQ_LENGTH, 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
    model.fit(X_train_lstm, y_train_lstm, epochs=20, batch_size=64, validation_split=0.2, verbose=0)

    predicted_prices_lstm = model.predict(X_test_lstm)
    predicted_prices_lstm = scaler.inverse_transform(predicted_prices_lstm)
    y_test_lstm = scaler.inverse_transform(y_test_lstm.reshape(-1, 1))

    print_prediction_accuracy(y_test_lstm, predicted_prices_lstm, 'LSTM', stock_name)

    simulate_trading(test_dates[window_size:], test_data[window_size:], predicted_prices_knn, f'KNN on {stock_name}')
    simulate_trading(test_dates[window_size:], test_data[window_size:], predicted_prices_dqn, f'DQN on {stock_name}')
    simulate_trading(stock_data['Date'][len(stock_data)-len(y_test_lstm):], y_test_lstm.flatten(), predicted_prices_lstm.flatten(), f'LSTM on {stock_name}')

for stock in stocks:
    stock_data = data[data['Name'] == stock]
    process_stock(stock_data, stock)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from datetime import timedelta

data = pd.read_csv('all_stocks_2006-01-01_to_2018-01-01.csv')

stock_data = data[data['Name'] == 'AAPL']

stock_data = stock_data.sort_values('Date')

stock_data['Date'] = pd.to_datetime(stock_data['Date'])

closing_prices = stock_data['Close'].values

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(closing_prices.reshape(-1, 1))

SEQ_LENGTH = 100

def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_step - 1):
        a = dataset[i:(i+time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)

train_data = scaled_data

X_train, y_train = create_dataset(train_data, SEQ_LENGTH)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)

def create_dqn_model(input_shape):
    model = Sequential()
    model.add(Dense(256, input_shape=input_shape, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1))  
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
    return model

input_shape = (SEQ_LENGTH, 1)
dqn_model = create_dqn_model(input_shape)

dqn_model.fit(X_train, y_train, epochs=20, batch_size=64, validation_split=0.2, verbose=0)

def predict_future_prices_dqn(model, data, scaler, seq_length, n_steps):
    future_predictions = []
    last_sequence = data[-seq_length:]

    for _ in range(n_steps):
        input_data = last_sequence.reshape((1, seq_length, 1))
        next_prediction = model.predict(input_data, verbose=0)
        future_predictions.append(next_prediction[0, 0])
        last_sequence = np.append(last_sequence[1:], next_prediction)
        last_sequence = last_sequence[-seq_length:] 

    future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
    return future_predictions

n_days_to_predict = (pd.to_datetime('2020-01-01') - stock_data['Date'].max()).days
predicted_future_prices_dqn = predict_future_prices_dqn(dqn_model, scaled_data, scaler, SEQ_LENGTH, n_days_to_predict)
last_date = stock_data['Date'].max()
future_dates = pd.date_range(start=last_date + timedelta(days=1), periods=n_days_to_predict, freq='D')

all_dates = np.concatenate((stock_data['Date'].values, future_dates))
all_prices = np.concatenate((closing_prices, predicted_future_prices_dqn.reshape(-1)))

initial_balance = 10000 
balance = initial_balance
shares = 0  

for i in range(len(stock_data) - SEQ_LENGTH - 1, len(all_prices) - 1):
    current_price = all_prices[i]
    next_price = all_prices[i + 1]

    
    if next_price > current_price and balance > 0:
        shares = balance / current_price
        balance = 0
        print(f"Bought shares at {current_price} on {all_dates[i]}")

    elif next_price < current_price and shares > 0:
        balance = shares * current_price
        shares = 0
        print(f"Sold shares at {current_price} on {all_dates[i]}")

final_balance = balance + shares * all_prices[-1]

print(f'Initial Balance: {initial_balance}')
print(f'Final Balance: {final_balance}')

plt.figure(figsize=(10, 6))
plt.plot(all_dates, all_prices, label='Actual and Predicted Prices')
plt.axvline(x=pd.to_datetime('2020-01-01'), color='red', linestyle='--', label='2020-01-01')
plt.title('Trading Simulation with DQN Predictions')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import norm
import math

def black_scholes(S, K, T, r, sigma, option_type='call'):
    """
    Calculate the Black-Scholes price for a European call or put option.

    Parameters:
    S : float : Stock price
    K : float : Strike price
    T : float : Time to maturity (in years)
    r : float : Risk-free interest rate (annual rate)
    sigma : float : Volatility of the underlying stock (annual standard deviation)
    option_type : str : 'call' or 'put'

    Returns:
    price : float : Price of the option
    """
    try:
        d1 = (math.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * math.sqrt(T))
        d2 = d1 - sigma * math.sqrt(T)
    except: return np.nan
    if option_type == 'call':
        try:
            price = S * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
        except: price=np.nan
    elif option_type == 'put':
        try:
            price = K * math.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
        except:price=np.nan
    return price
```

### Efficient Market Hypothesis (EMH)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

file_path = '/content/all_stocks_2006-01-01_to_2018-01-01.csv'
data = pd.read_csv(file_path)

data['Date'] = pd.to_datetime(data['Date'])

stocks = ['AAPL', 'GOOGL', 'PG']

def calculate_returns(stock_data):
    stock_data['Return'] = stock_data['Close'].pct_change()
    return stock_data

def perform_adf_test(series):
    result = adfuller(series.dropna())
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

def calculate_and_plot_volatility(stock_data, stock_name):
    stock_data['RollingVolatility'] = stock_data['Return'].rolling(window=252).std() * np.sqrt(252)
    plt.figure(figsize=(14, 7))
    plt.plot(stock_data['Date'], stock_data['RollingVolatility'], label=f'{stock_name} Rolling Volatility')
    plt.title(f'Rolling Volatility of {stock_name}')
    plt.xlabel('Date')
    plt.ylabel('Volatility')
    plt.legend()
    plt.show()

for stock in stocks:
    stock_data = data[data['Name'] == stock]
    stock_data = calculate_returns(stock_data)

    print(f'ADF Test for {stock} Returns:')
    perform_adf_test(stock_data['Return'])
    calculate_and_plot_volatility(stock_data, stock)
```

The provided images display the rolling volatility of AAPL, PG, and GOOGL, showing significant spikes during the 2008 recession followed by stabilization in the subsequent years. This trend supports the Efficient Market Hypothesis (EMH), which posits that stock prices fully reflect all available information. The sharp increase in volatility across all three stocks during the financial crisis demonstrates the market's rapid reaction to new, adverse economic conditions. The gradual decrease and stabilization of volatility post-recession indicate that the market efficiently processed and incorporated this information, adjusting stock prices to reflect the new economic realities. The similar volatility patterns across different sectors (technology for AAPL and GOOGL, consumer goods for PG) further suggest that the market responded efficiently to the widespread impact of the financial crisis.

Another limitation is the use of default parameters for our models, which limits their performance. Specifically, the DQN model requires more hyperparameter tuning to make accurate predictions. This likely contributed to its struggles in our trading simulations. Additionally, these models are probably unable to handle sudden market changes, such as political shifts or severe economic crises.

### Ethics & Privacy

There are ethical concerns regarding the privacy and security of the data used in this project. Since we are using historical financial data, there may be privacy concerns or regulations related to the companies from which this information is sourced. In the finance world, various standards exist to control ethical conduct. This tool's predictions should not be used to influence stock prices by publicly posting the results. Users of the model should be aware of its limitations, as it is not a fully comprehensive and accurate predictor of stock prices.

### Conclusion

In this project, we aimed to determine the effectiveness of different machine learning models (KNN, DQN, and LSTM) in predicting stock prices and evaluated their performance using RMSE and MAPE. Our results show that the LSTM model consistently outperforms KNN and DQN models across AAPL, GOOGL, and PG stocks, indicating it is the most accurate model for stock price prediction in our study. The LSTM model's lower RMSE and MAPE values demonstrate its superior ability to capture long-term dependencies and complex patterns in stock price data. Additionally, the consistent performance of LSTM across different types of stocks highlights its robustness and generalizability. By simulating trading based on model predictions, we illustrated how these models can inform trading decisions, while volatility analysis provided insights into market behavior and the impact of market efficiency on stock prices. Our findings align with existing research supporting the effectiveness of deep learning models, particularly LSTMs, in financial forecasting. Future work can explore incorporating additional financial indicators, developing hybrid models, implementing real-time data applications, and extending prediction horizons to enhance predictive accuracy and practical applicability in live markets. This project underscores the potential of machine learning models in financial forecasting and lays a foundation for further exploration in stock market prediction techniques.
